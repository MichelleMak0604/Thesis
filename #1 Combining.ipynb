```python
#make lists of the column names of each dataframe
wave1and2_colnames = wave1and2.columns
wave3_colnames = wave3.columns

#same column names
common_cols = wave1and2_colnames.intersection(wave3_colnames)
print("amount of columns that are the same:", len(common_cols))

#column names in wave1and2, but not in wave3
different_wave1and2 = wave1and2_colnames.difference(wave3_colnames)
print("amount of columns that occur in wave1and2, but not in wave3:", len(different_wave1and2))

#column names in wave3, but not in wave1and2
different_wave3 = wave3_colnames.difference(wave1and2_colnames)
print("amount of columns that occur in wave3, but not in wave1and2:", len(different_wave3))
```

```python
#make a new dataframes with only the columns that are common 
wave1and2_common = wave1and2[[*common_cols]]
wave3_common = wave3[[*common_cols]]
```

```python
#combine both new dataframes into 1 dataframe: waves
waves = [wave1and2_common, wave3_common]
waves = pd.concat(waves)
```

```python
#info about the new and final dataframe
waves.info()
```

```python
#see the values for the Covid result
waves.Covid_result.value_counts()
```

```python
#delete all the rows where the OD.CO value is NaN (no value is measured)
waves = waves[waves['OD.CO'].notna()]

#see the values for the Covid result
waves.Covid_result.value_counts()
```
